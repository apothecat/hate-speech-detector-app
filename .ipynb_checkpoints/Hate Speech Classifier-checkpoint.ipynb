# Hate Speech Classifier

```python
# Read a comma-separated values (csv) file into DataFrame
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html

import pandas as pd

train_en = pd.read_csv("data/train/hateval2019_en_train.csv")

# import stopwords

import nltk
nltk.download('stopwords')

# display dataframe

train_en.head()
```

```python
import re
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
import string

#train_en.columns = ['id', 'text', 'HS', 'TR', 'AG']
train_en = train_en[0:500]
train_en.head()

# Create feature (percentage capital letter in tweet)

import string

def count_caps(text):
    count = sum(1 for char in text if char.isupper())
    return round(count/(len(text) - text.count(" ")), 3)*100

train_en['caps%'] = train_en['text'].apply(lambda x: count_caps(x))


# Clean Text

def clean_text(text):
    punct_string = string.punctuation.replace("#","") 
    stopword = nltk.corpus.stopwords.words('english')
    text = "".join([word for word in text if word not in punct_string])
    tokens = re.split('\W+', text)
    text = [word.lower() for word in tokens if word not in stopword]
    return text

# TF-IDF
tfidf_vect = TfidfVectorizer(analyzer=clean_text)
X_tfidf = tfidf_vect.fit_transform(train_en['text'])
X_tfidf_feat = pd.concat([train_en['caps%'], pd.DataFrame(X_tfidf.toarray())], axis=1)

# CountVectorizer
#count_vect = CountVectorizer(analyzer=clean_text)
#X_count = count_vect.fit_transform(train_en['text'])
#X_count_feat = pd.concat([train_en['caps%'], pd.DataFrame(X_count.toarray())], axis=1)

X_tfidf_feat.head()
```

```python
from sklearn.metrics import precision_recall_fscore_support as score
from sklearn.model_selection import train_test_split

# Create training set and test set
X_train, X_test, y_train, y_test = train_test_split(X_tfidf_feat, train_en['HS'], test_size=0.2)
```

```python
rf = RandomForestClassifier(n_jobs=-1)
#param = {'n_estimators': [10, 150, 300],
#        'max_depth': [30, 60, 90, None]}
#gs = GridSearchCV(rf, param, cv=5, n_jobs=-1)

# Feed in training data to fit the model, across the different folds, in each of the parameter settings

#gs_fit = gs.fit(X_tfidf_feat, train_en['HS'])
hsmod_clf = rf.fit(X_train, y_train)
# predict

y_pred = hsmod_clf.predict(X_test)
precision, recall, fscore, support = score(y_test, y_pred, average = 'micro')

#pd.DataFrame(hsmod_clf.cv_results_).sort_values('mean_test_score', ascending=False)[0:5]
pd.DataFrame(hsmod_clf.cv_results_)


```

```python
# Saving knn_clf
import joblib
# Save the model as a pickle in a file
joblib.dump(hsmod_clf, "hsmod.pkl")
```

## Example

```python
# See nlp_training notebook for example of a final Random Forest Model

import pandas as pd

# Read in data
train_en = pd.read_csv("data/train/hateval2019_en_train.csv")
train_en = train_en.drop(columns = ['TR', 'AG'])
# train_en

# Store features matrix in X
X= train_en.text
#Store target vector in y
y= train_en.HS
# Will eventually be final model after evaluation and choosing best parameter
# Currently just using a simple example
# 
# Import Classifier
from sklearn.ensemble import RandomForestClassifier
# instantiate the model with the best known parameters
hsmod = RandomForestClassifier(n_jobs=-1)
# train the model with X and y (not X_train and y_train)
hsmod_clf=hsmod.fit(X, y)
# Saving knn_clf
import joblib
# Save the model as a pickle in a file
joblib.dump(hsmod_clf, "hsmod.pkl")
```

```python
from sklearn.datasets import load_iris
iris= load_iris()
# Store features matrix in X
X= iris.data
#Store target vector in y
y= iris.target
# Finalizing KNN Classifier after evaluation and choosing best 
# parameter
#Importing KNN Classifier
from sklearn.neighbors import KNeighborsClassifier
# instantiate the model with the best known parameters
knn = KNeighborsClassifier(n_neighbors=12)
# train the model with X and y (not X_train and y_train)
knn_clf=knn.fit(X, y)
# Saving knn_clf
import joblib
# Save the model as a pickle in a file
joblib.dump(knn_clf, "Knn_Classifier.pkl")
```
